Import the libraraies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Pre- Processing

```python
# loading the data from the csv file
raw_mail_data = pd.read_csv('/content/mail_data.csv')
raw_mail_data
```

```python
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), '')
```

```python
# printing the first 5 rows of the dataframe
mail_data.head()
```

```python
# checking the no of rows and columns in the dataframe
mail_data.shape
```

Label Encoding

```python
# label spam mail as 0; ham mail as 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category',]=0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',]=1
```

Spam --> 0

Ham -->1

```python
# separating the data as text and label
x = mail_data['Message']
y = mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into Training Data and Test Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Feature Extraction


```python
# tansform the text data to feature vectors that can be used as input to the logistic regression
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True) # Change lowercase='True' to lowercase=True

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

# convert y_train and y_test values as integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features)
```

Training logistics regression model

```python
model = LogisticRegression()
```

```python
# training logistic regression model with the training data
model.fit(x_train_features,y_train)

```

Evaluating the trained model


```python
# prediction on training data
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data :',accuracy_on_training_data )
```

```python
# prediction for testing data
prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test,prediction_on_test_data)
```

```python
print('Accuracy on test data:',accuracy_on_test_data)
```

Building a Predictive System

```python
input =["Ok lar... Joking wif u oni..."]

# convert text to features vectors
input_features = feature_extraction.transform(input)

# making prediction
prediction = model.predict(input_features)
print(prediction)

if (prediction[0]==1):
  print('Ham mail')
else:
  print('Spam mail')
```
